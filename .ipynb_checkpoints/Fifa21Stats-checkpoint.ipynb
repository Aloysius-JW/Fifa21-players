# Looking through data and their meaning
The data is from [sofifa.com](sofifa.com) and looking through how the data is represented.
- photoUrl and PlayerUrl = hyperlinks when they are scrapped for the website
- LongName and Name = Names of player
- Nationality = Nationality
- Position = the postions that they can play in
- Team & Contract = Their team and how long their contract last/if the even have a contract
- OVA and BOV = Best Overall
- POT = Potential
- Growth = Potential - Best Overall
- Weight = Weight
- Foot = Their preferred foot/ dominant foot
- BP = Best Position
- Joined = Date the player joined their current club
- Loan Date End = whether the player is on loan and if he is, when the loan is expiring
- Value = Market value
- Wages = Amount to get paid per week according to contract
- Release Clause = amount of money to be paid so that the club is obliged to accept
- Attacking = Crossing+Finishing+Heading+Accuracy+Short+Passing+Volleys	
- Skill =Dribbling+Curve+FK Accuracy+Long Passing+Ball Control	
- Movement = Acceleration+Sprint Speed+Agility+Reactions+Balance
- Power =  Shot Power+Jumping+Stamina+Strength+Long Shots
- Mentality = Aggression+Interceptions+Positioning+Vision+Penalties
- Defending=Marking+Standing Tackle+Sliding Tackle
- Goalkeeping= GK Diving+GK Handling+GK Kicking+GK Positioning+GK Reflexes

- W/F = Weak Foot ranking out of 5
- S/M = Skill Moves ranking out of 5
- A/W = Attacking Work Rate low,medium,High
- D/W = Defensive Work Rate low,medium,High
- IR = International Reputation
- PAC = Pace/Diving
- SHO = Shooting/Handling
- PAS = Passing/Kicking
- DRI = Dribbling/Reflexes
- DEF = Defending/Pace
- PHY = Physical/Positioning
- Hits = ? What i deduce is maybe the number of people who search for this particular person. Maybe useless in this case

```python
#start by dropping columns that i think its unnecessary 
#url for pictures and player are some of them
#looks like the short form name is not very useful either
df.columns
df = df[[#'photoUrl',
        'ID',
        'LongName', 
        #'playerUrl', 
        'Nationality', 'Positions', 
        #'Name',
       'Age', '↓OVA', 'POT', 'Team & Contract',  'Height', 'Weight',
       'foot', 'BOV', 'BP', 'Growth', 'Joined', 'Loan Date End', 'Value',
       'Wage', 'Release Clause', 'Attacking', 'Crossing', 'Finishing',
       'Heading Accuracy', 'Short Passing', 'Volleys', 'Skill', 'Dribbling',
       'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control', 'Movement',
       'Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance',
       'Power', 'Shot Power', 'Jumping', 'Stamina', 'Strength', 'Long Shots',
       'Mentality', 'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Penalties', 'Composure', 'Defending', 'Marking', 'Standing Tackle',
       'Sliding Tackle', 'Goalkeeping', 'GK Diving', 'GK Handling',
       'GK Kicking', 'GK Positioning', 'GK Reflexes', 'Total Stats',
       'Base Stats', 'W/F', 'SM', 'A/W', 'D/W', 'IR', 'PAC', 'SHO', 'PAS',
       'DRI', 'DEF', 'PHY', 'Hits'
        ]]
```

```python
# renaming columns to get full names
# then correcting individual columns
df = df.rename(columns={'↓OVA' : 'OVA',
                        'LongName' : 'Name',
                        'Loan Date End' : 'Loan_Date_End',
                        'Release Clause' : 'Release_Clause',
                        'Heading Accuracy': 'Heading_Accuracy',
                        'Short Passing' : 'Short_Passing',
                        'FK Accuracy' : 'FK_Accuracy',
                        'Long Passing': 'Long_Passing',
                        'Ball Control': 'Ball_Control',
                        'Sprint Speed': 'Sprint_Speed',
                        'Shot Power': 'Shot_Power',
                        'Long Shots': 'Long_Shots',
                        'Standing Tackle' : 'Standing_Tackle',
                        'Sliding Tackle' : 'Sliding_Tackle',
                        'GK Diving' : 'GK_Diving',
                        'GK Handling' : 'GK_Handling',
                        'GK Kicking' : 'GK_Kicking',
                        'GK Positioning' : 'GK_Positioning',
                        'GK Reflexes' : 'GK_Reflexes',
                        'Total Stats' : 'Total_Stats',
                        'Base Stats' : 'Base_Stats'
                       })
```

```python
#splitting the columns to foot and inch and then doing a calculation to change foot and inches to cm in Height
# renaming the height to Height_in_cm 
df[['foot','inch']] = df['Height'].str.split("'", expand = True)
df['inch'] = df['inch'].str.strip('"')
df['foot'] = pd.to_numeric(df['foot'],errors='raise')
df['inch'] = pd.to_numeric(df['inch'],errors='coerce')
df['Height'] = df['foot'] * 30.48 + df['inch'] * 2.54 
df = df.rename(columns = {'Height' : 'Height_in_cm'})
df = df.drop(columns=['foot', 'inch'])
# df.head(2)
```

```python
# getting rid of the 'lbs' in weight and converting lbs to KG
# renaming weight header as well
df['Weight'] = df['Weight'].str.strip('lbs')
df = df.rename(columns = {'Weight' : 'Weight_in_kg'})
df['Weight_in_kg'] = pd.to_numeric(df['Weight_in_kg'])
df['Weight_in_kg'] = df['Weight_in_kg'] * 0.453592
df['Weight_in_kg'] = round(df['Weight_in_kg'],2)
```

```python
#Creating a function to clean up money to usable format
def money(x):
#if it's already transformed, do nothing
    if isinstance(x,float):
        return x
    if isinstance(x,int):
        return x
#get rid of the euro sign first
    x = x.strip('€')
#if with an M, get the number to multiple by a million
    if 'M' in x:
        x = x.strip('M')
        x = float(x)* 1000000
#if with an K, get the number to multiple by a thousand
    elif 'K' in x:
        x = x.strip('K')
        x = float(x) * 1000
    return int(x)

df['Value'] = df['Value'].apply(money)
df['Wage'] = df['Wage'].apply(money)
df['Release_Clause'] = df['Release_Clause'].apply(money)

df['Value'] = pd.to_numeric(df['Value'])
df['Wage'] = pd.to_numeric(df['Wage'])
df['Release_Clause'] = pd.to_numeric(df['Release_Clause'])
```

```python
#getting rid of the stars from W/F, SM and IR
df['W/F'] = pd.to_numeric( df['W/F'].str.strip(' ★'))
df['SM'] = pd.to_numeric(df['SM'].str.strip(' ★'))
df['IR'] = pd.to_numeric(df['IR'].str.strip(' ★'))
```

```python
#getting rid of the \n in Hits
df['Hits'] = df['Hits'].str.strip('\n')
```

```python
# starting from the team and contract columns
df["Team & Contract"]=df["Team & Contract"].str.strip("\n")
df[['Team', 'Contract']] = df['Team & Contract'].str.split('\n', expand=True)
df[['Year_Contract_Start', 'Year_Contract_End']] = df['Contract'].str.split('~', expand=True)

# df['Contract_End'].value_counts()
# df['Contract_Start'].value_counts()
# df['Team'].value_counts()

## Saw some teams have '1.' infront of their team name. Some have whitespace in front of their name as well. getting rid of them
df['Team'] = df['Team'].str.lstrip()
df['Team'] = df['Team'].str.replace("1.",'')
df[df['Team'].str.match("Norway")]

## upon closer inspection, there are players that do not have teams. In this case, their Contract columns will indicate 'Free',
## therefore, will be replacing the value to Free for the Contract_Start and Contract_end conlumns 
## Same thing for the 'On loan' players, while this time, i'll change the value to On Loan in the Contract_Start and Contract_end columns
def Copying_Free(x,y):
    if x=='Free':
        return x
    else:
        return y

def On_Loan(x,y):
    if 'On Loan' in x:
        return 'On Loan'
    else:
        return y

df['Year_Contract_End']=df.apply(lambda x: Copying_Free(x['Contract'] , x['Year_Contract_End']), axis = 1)
df['Year_Contract_Start']=df.apply(lambda x: Copying_Free(x['Contract'] , x['Year_Contract_End']), axis = 1)

df['Year_Contract_End']=df.apply(lambda x: On_Loan(x['Contract'] , x['Year_Contract_End']), axis = 1)
df['Year_Contract_Start']=df.apply(lambda x: On_Loan(x['Contract'] , x['Year_Contract_End']), axis = 1)

# Also filling up the Null values in Loan_Date_End as "Not on loan'
def On_Loan_Date(x):
    if 'On Loan' in x:
        return x
    else:
        return 'Not on loan'

df['Loan_Date_End']=df.apply(lambda x: On_Loan_Date(x['Contract']), axis = 1)

```
